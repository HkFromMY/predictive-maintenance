# Exploratory Data Analysis (EDA)

```python
def plot_pie_dist(labels, values, title):
    plt.pie(x=values, labels=labels, autopct="%.2f")
    plt.title(title)

    plt.show()
    
def box_strip_dist(x, y):
    plt.subplot(1, 2, 1)
    sns.boxplot(x=x, y=y)

    plt.subplot(1, 2, 2)
    sns.stripplot(x=x, y=y)

    plt.tight_layout()
    plt.show()
    
def comparative_graph(x, y):
    contingency_table = pd.crosstab(index=x, columns=y)
    contingency_table.plot(kind='bar', rot=0)
    plt.show()
    
    print("Contingency table:")
    return contingency_table
```

# Data Pre-Processing

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
class Preprocessor:
    def __init__(self):
        self.encoder = LabelEncoder()
        self.scaler = MinMaxScaler(feature_range=(0, 1))
        self.numerical_variables = [
            'Air temperature [K]',
            'Process temperature [K]',
            'Rotational speed [rpm]',
            'Torque [Nm]',
            'Tool wear [min]',
            'power',
            'tool_wear_rate',
            'torque_wear_product',
            'temperature_diff'
        ]
    
    def generate_features(self, data):
        """
        This function generate new features for the training and testing data
        source: https://www.kaggle.com/code/yogesh239/lgbm-optuna-predict-machine-failure-probability 
        """
        data['power'] = data['Torque [Nm]'] * data['Rotational speed [rpm]']
        data['tool_wear_rate'] = data['Tool wear [min]'] / data['Tool wear [min]'].max()
        data['total_failures'] = data[['TWF', 'HDF', 'PWF', 'OSF', 'RNF']].sum(axis=1)
        data['torque_wear_product'] = data['Torque [Nm]'] * data['Tool wear [min]']
        data['temperature_diff'] = np.abs(data['Process temperature [K]'] - data['Air temperature [K]'])
        
        return data
    
    def preprocess_train(self, data):
        """
        This function preprocess the training set
        """
        data = self.generate_features(data)
        data['Type'] = self.encoder.fit_transform(data['Type'])
        data[self.numerical_variables] = self.scaler.fit_transform(data[self.numerical_variables])
        
        return data
    
    def preprocess_test(self, data):
        """
        This function preprocess the testing set
        """
        data = self.generate_features(data)
        data['Type'] = self.encoder.transform(data['Type'])
        data[self.numerical_variables] = self.scaler.transform(data[self.numerical_variables])
        
        return data
        
```

```python
preprocessor = Preprocessor()
train_data = preprocessor.preprocess_train(train_df)

test_data = preprocessor.preprocess_test(test_df)
```

```python
new_columns = {
    "Air temperature [K]": "Air temperature",
    "Process temperature [K]": "Process temperature",
    "Rotational speed [rpm]": "Rotational speed",
    "Torque [Nm]": "Torque",
    "Tool wear [min]": "Tool wear"
}

train_data = train_data.rename(columns=new_columns)
test_data = test_data.rename(columns=new_columns)
```

# Model Development

```python
import xgboost as xgb
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import RidgeClassifier, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
from scipy import stats
from sklearn.metrics import roc_auc_score, roc_curve

import tensorflow as tf
```

```python
class Model:
    def __init__(self, name):
        self.model_name = name
        
    def confusion_plot(self, y_true, y_pred):
        """
        This function visualizes the confusion matrix with heatmap
        """

        matrix = confusion_matrix(y_true, y_pred)
        tn, fp, fn, tp = matrix.ravel()

        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f1_score = (2 * precision * recall) / (precision + recall)

        print("Precision:", precision)
        print("Recall:", recall)
        print("F1 Score:", f1_score)

        values = ["{0:0.0f}".format(value) for value in matrix.flatten()]
        percentages = ["{0:.2%}".format(value) for value in matrix.flatten() / np.sum(matrix)]

        labels = ['True Negative', 'False Positive', 'False Negative', 'True Positive']
        labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(labels, values, percentages)]
        labels = np.asarray(labels).reshape(2, 2)

        sns.heatmap(matrix, annot=labels, fmt="", cmap="rocket_r")
        plt.title(f"Confusion Matrix for {self.model_name} Model")

        plt.show()
        
    def auc_plot(self, y_test, y_pred):
        """
        This function visualize the ROC-AUC plot
        """
        fpr, tpr, _ = roc_curve(y_test, y_pred)
        auc = roc_auc_score(y_test, y_pred)

        plt.plot(fpr, tpr, label="AUC = " + str(round(auc, 4)))
        plt.ylabel("True positive rate")
        plt.xlabel("False positive rate")
        plt.legend()
        plt.show()
        
class NeuralNetwork(Model):
    def __init__(self):
        """
        Create a neural network
        """
        tf.random.set_seed(42)
        super().__init__("Neural Network")
        
        self.model = tf.keras.Sequential([
            tf.keras.layers.Dense(
                512,
                activation="relu"
            ),
            tf.keras.layers.Dense(
                512,
                activation="relu"
            ),
            tf.keras.layers.Dense(
                1,
                activation="sigmoid"
            )
        ])

        self.model.compile(
            loss=tf.keras.losses.BinaryCrossentropy(),
            optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
            metrics=["accuracy"]
        )
        
    def train(self, X_train, y_train, epochs=None):
        """
        Train the model with data according to the epochs set
        """
        if epochs is None:
            epochs = 50 # default
            
        earlystop = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=2)
        self.history = self.model.fit(
            X_train, 
            y_train, 
            epochs=epochs, 
            batch_size=32, 
            validation_split=0.2, 
            callbacks=[earlystop]
        )
        
        return self.history
    
    def predict(self, X_test):
        """
        Make predictions and convert to Pandas Series object
        """
        pred = self.model.predict(X_test)
        pred = pred.flatten()
        pred = pd.Series(pred)
        
        return pred
    
    def get_summary(self):
        return self.model.summary()
    
class Ridge(Model):
    def __init__(self):
        super().__init__("Ridge")
        self.model = RidgeClassifier()
        
    def train(self, X_train, y_train):
        self.model.fit(X_train, y_train)
        
    def predict(self, X_test):
        return self.model.predict(X_test)
    
class RandomForest(Model):
    def __init__(self, **kwargs):
        super().__init__("Random Forest")
        self.model = RandomForestClassifier(random_state=42, **kwargs)
        
    def train(self, X_train, y_train):
        self.feature_names = X_train.columns
        self.model.fit(X_train, y_train)
        
    def predict(self, X_test):
        return self.model.predict(X_test)
    
    def feature_importances_plot(self):
        feature_importances = np.vstack([self.feature_names, self.model.feature_importances_])
        feature_importances = pd.DataFrame(feature_importances).T.sort_values(by=1, ascending=False) # sort by feature importances
        
        # plotting
        plt.figure(figsize=(16, 9))
        sns.barplot(data=feature_importances, x=1, y=0, orient='h', color='steelblue')
        plt.title("Feature Importances for Random Forest Model")
        plt.xlabel("")
        plt.ylabel("")
        plt.show()
    
class XGBoost(Model):
    def __init__(self, **kwargs):
        super().__init__("XGBoost")
        self.model = xgb.XGBClassifier(scale_pos_weight=(107433 / 1710), **kwargs)
        
    def train(self, X_train, y_train):
        self.model.fit(X_train, y_train)
        
    def predict(self, X_test):
        return pd.Series(self.model.predict(X_test))
    
class LogisticModel(Model):
    def __init__(self):
        super().__init__("Logistic Regression")
        self.model = LogisticRegression()
        
    def train(self, X_train, y_train):
        self.model.fit(X_train, y_train)
        
    def predict(self, X_test):
        return self.model.predict(X_test)
    
    def predict_proba(self, X_test):
        return self.model.predict_proba(X_test)[:, 1] # second column which is the probability for class 1
    
class LGBM(Model):
    def __init__(self, **kwargs):
        super().__init__("Light GBM")
        self.model = LGBMClassifier(**kwargs)
        
    def train(self, X_train, y_train):
        self.model.fit(X_train, y_train)
        
    def predict(self, X_test):
        return self.model.predict(X_test)
    
class Hybrid(Model):
    
    def __init__(self, models):
        super().__init__("Hybrid")
        self.models = models
        
    def train(self, X_train, y_train):
        for model in self.models:
            model.train(X_train, y_train)
            print(f"Training complete for {model.model_name}")
            
    def predict(self, X_test):
        preds = []
        
        # each model making predictions
        for model in self.models:
            pred = model.predict(X_test)
            
            if model.model_name == "Neural Network":
                pred = pred.round(0)
                
            preds.append(pred)
        
        # find mode of each row of predictions by all models (max-voting technique)
        final_pred = stats.mode(preds, keepdims=False).mode
        
        return final_pred
```

```python
# train models
rc = Ridge()
rc.train(X_train, y_train)

rf = RandomForest(n_estimators=100, class_weight="balanced")
rf.train(X_train, y_train)

xgb_clf = XGBoost(n_estimators=150, max_depth=6)
xgb_clf.train(X_train, y_train)

lg_clf = LogisticModel()
lg_clf.train(X_train, y_train)

lgbm_clf = LGBM()
lgbm_clf.train(X_train, y_train)

nn_clf = NeuralNetwork()
nn_clf.train(X_train, y_train, epochs=50)

# predict
rc_pred = rc.predict(X_test)
rf_pred = rf.predict(X_test)
xgb_pred = xgb_clf.predict(X_test)
lg_pred = lg_clf.predict(X_test)
lgbm_pred = lgbm_clf.predict(X_test)
nn_pred = nn_clf.predict(X_test)
```

```python
hybrid = Hybrid([
    Ridge(),
    RandomForest(n_estimators=100, class_weight="balanced"),
    XGBoost(n_estimators=150, max_depth=6),
    LogisticModel(),
    LGBM(),
    NeuralNetwork()
])

hybrid.train(X_train, y_train)
hybrid_pred = hybrid.predict(X_test)
```
